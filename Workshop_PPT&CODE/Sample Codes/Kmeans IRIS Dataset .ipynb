```python
from sklearn import datasets
from sklearn.cluster import KMeans
```

```python
# Load IRIS dataset

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
# Instantiate the KMeans models
#
km = KMeans(n_clusters=3, random_state=42)
```

```python
#
# Fit the KMeans model
#
km.fit_predict(X)
```

```python
# Calculate Silhoutte Score
from sklearn.metrics import silhouette_score 
score = silhouette_score(X, km.labels_, metric='euclidean')
```

```python
# Print the score

print('Silhouetter Score: %.3f' %score)
```

```python
# show silhouette_score with cluster 
# from output we found that 5Cluster,Score is the maximum score 
# so we can finalize the cluster size =5 
k=range(2,15)
sse=[]
for i in k:
    demo_model=KMeans(n_clusters=i,random_state=0).fit(X)
    sse.append(demo_model.inertia_)
    y=demo_model.predict(X)
    print(f"{i}Cluster,Score{silhouette_score(X,y)}")
plt.scatter(k,sse)
plt.plot(k,sse)
```

#Yellowbrick extends the Scikit-Learn API to make model selection and hyperparameter tuning easier. It provides some very useful wrappers to create the visualisation in no time. Here is the code to create Silhouette plot for K-Means clusters with n_cluster as 2, 3, 4, 5.

```python
#https://www.scikit-yb.org/en/latest/quickstart.html#installation
#https://www.scikit-yb.org/en/latest/api/cluster/silhouette.html
#pip install yellowbrick
#conda install -c districtdatalabs yellowbrick

from yellowbrick.cluster import SilhouetteVisualizer
import matplotlib.pyplot as plt
 
fig, ax = plt.subplots(2, 2, figsize=(10,8))
for i in [2, 3, 4, 5]:
    '''
    Create KMeans instance for different number of clusters
    '''
    km = KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=100, random_state=42)
    q, mod = divmod(i, 2)
    '''
    Create SilhouetteVisualizer instance with KMeans instance
    Fit the visualizer
    '''
    visualizer = SilhouetteVisualizer(km, colors='yellowbrick', ax=ax[q-1][mod])
    visualizer.fit(X) 
```
